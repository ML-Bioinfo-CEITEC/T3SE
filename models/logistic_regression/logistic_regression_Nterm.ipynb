## Load the training data:

```python
import pandas as pd 

df_train = pd.read_csv('/home/jovyan/data/secreted_proteins/T3SE_clustered_embeddings_Nterm_train.csv')
df_train
```

```python
df_train.drop(columns=['ID', 'Sequence_part'], inplace=True)
df_train['Label'].value_counts()
```

```python
X_train = df_train.iloc[:,1:]
X_train
```

```python
from sklearn import preprocessing


def normalise(df):
    np_arr = df.values # returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    scaled_data = min_max_scaler.fit_transform(np_arr)
    return pd.DataFrame(scaled_data)


# def standardise(df):
#     np_arr = df.values
#     standard_scaler = preprocessing.StandardScaler()
#     scaled_data = standard_scaler.fit_transform(np_arr)
#     return pd.DataFrame(scaled_data)


X_train = normalise(X_train)
#X_train = standardise(X_train)
X_train
```

```python
Y_train = df_train['Label']
Y_train
```

## Load the testing data:

```python
df_test = pd.read_csv('/home/jovyan/data/secreted_proteins/T3SE_clustered_embeddings_Nterm_test.csv')
df_test
```

```python
df_test['Label'].value_counts()
```

```python
df_test.drop(columns=['ID', 'Sequence_part'], inplace=True)
X_test = df_test.iloc[:,1:]
X_test = normalise(X_test)
#X_test = standardise(X_test)
Y_test = df_test['Label']
```

## Fit logistic regression:

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

# model = LogisticRegression(class_weight={0:1, 1:1}, max_iter=500).fit(X_train, Y_train)  # 0.9310 on test (27 positives correct, 48 positives as negative, 4 negative as positive)
# model = LogisticRegression(class_weight={0:1, 1:5}, max_iter=500).fit(X_train, Y_train)  # 0.9297 on test (36 positive correct, 39 positives as negative, 14 negative as positive)
# model = LogisticRegression(class_weight={0:1, 1:9}, max_iter=500).fit(X_train, Y_train)  # 0.9257 on test (40, 35, 21)
# model = LogisticRegression(class_weight={0:1, 1:10}, max_iter=500).fit(X_train, Y_train) # 0.9244 on test (41, 34, 23)
# model = LogisticRegression(class_weight={0:1, 1:11}, max_iter=500).fit(X_train, Y_train) # 0.9244 on test (42, 33, 24)
# model = LogisticRegression(class_weight={0:1, 1:15}, max_iter=500).fit(X_train, Y_train) # 0.9191 on test (43, 32, 29)
model = LogisticRegression(class_weight='balanced', max_iter=500).fit(X_train, Y_train)  # 0.9178 on test (46, 29, 33)
model
```

```python
print(f'coefficient of determination: {model.score(X_train, Y_train)}')
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

```python
print(f'coefficient of determination: {model.score(X_test, Y_test)}')
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
from sklearn import metrics

print(f'Accuracy: {metrics.accuracy_score(Y_test, Y_pred)}')
print(f'F1: {metrics.f1_score(Y_test, Y_pred)}')

cm = metrics.confusion_matrix(Y_test, Y_pred)
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
Y_pred_proba = model.predict_proba(X_test)[::,1]
# fpr, tpr, _ = metrics.roc_curve(Y_test,  Y_pred_proba)
auc = metrics.roc_auc_score(Y_test, Y_pred_proba)
print(f'AUC: {auc}')
# plt.plot(fpr, tpr)
# plt.show()
```
